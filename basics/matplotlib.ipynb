```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# this line works only in jupyter notebook. without this line, we have a use "plt.show()" to show the plot 
%matplotlib inline
```

```python
x = np.arange(0, 10)
x
```

```python
y = x**2
y
```

```python
# (x values, y values, **kwargs)
plt.plot(x,y,'blue')
```

```python
plt.plot(x, y, 'r--')
# x and y value limiting
plt.xlim(0, 4)
plt.ylim(0, 10)
```

```python
plt.plot(x, y, '-')
# title and label
plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
mat = np.arange(0, 100).reshape(10, 10)
mat
```

```python
# show image using imshow function
plt.imshow(mat)
```

```python
# use different color or style using cmap value
plt.imshow(mat, cmap='coolwarm')
```

```python
mat = np.random.randint(0, 1000, (10, 10))
mat
```

```python
plt.imshow(mat)
# colorbar in the side
plt.colorbar()
```
